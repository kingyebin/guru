```python
# 정확한 형식의 테마 목록 정의
themes = [
    "4X (explore, expand, exploit, exterminate)", "Abstract", "Action", "Adult", "Adventure", "Alternate Historical", "Anime", "Aquatic",
    "Business", "Civil War", "Comedy", "Comic Book", "Crime", "Cyberpunk", "Dating", "Drama", "Educational", "Egyptian", "Espionage", "Erotic",
    "Fantasy", "Fighting", "Game Show", "Historical", "Horror", "Kids", "Management", "Martial Arts", "Mayan", "Medieval", "Modern Military", "Motorsports",
    "Mystery", "Non-fiction", "Open world", "Party", "Post-Apocalyptic", "Prehistoric", "Romance", "Sandbox", "Sci-Fi", "Simulation", "Sports", "Stealth",
    "Steampunk", "Strategy", "Superhero", "Survival", "Thriller", "Vietnam", "Visual Novel", "Warfare", "Western", "World War II", "Time Travel", "Urban Exploration",
    "Farm Management", "Space Exploration", "Zombie Survival", "Pirate Adventures", "Detective", "Investigation", "Mythology", "Dystopian Future",
    "Magic", "Sorcery", "Music", "Rhythm", "Nature", "Wildlife Conservation", "Alien Worlds", "Ancient Civilizations", "Underwater Exploration",
    "Political Intrigue", "Supernatural Phenomena", "VR", "Art and Design", "Culinary Arts", "Sports Management", "Epic Sagas",
    "Heist", "Robbery", "Hospital", "Healthcare", "Disaster Management", "Urban Development", "Intergalactic War", "Time Management",
    "Psychological Thriller", "Lovecraftian Horror", "Biographical", "Noir"
]

# 테마를 추출하고 theme 열 업데이트 함수
def update_themes(row):
    current_tags = [tag.strip() for genre in str(row['Keywords']).split(',') for tag in genre.split('/')]
    current_themes = str(row['theme']).split('||') if pd.notna(row.get('theme', None)) else []
    new_themes = current_themes + [theme for theme in themes if any(theme.lower() in tag.lower() for tag in current_tags) and theme not in current_themes]
    new_themes = sorted(set(new_themes))
    return '||'.join(new_themes) if new_themes else None

# 'theme' 열이 없으면 추가
if 'theme' not in filtered_developers.columns:
    filtered_developers['theme'] = None

# 장르 열 업데이트
filtered_developers['theme'] = filtered_developers.apply(update_themes, axis=1)
filtered_developers
```
