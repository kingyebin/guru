```python
# 날짜 문자열을 표준 날짜 형식으로 변환하는 함수
def convert_date(date_str):
    try:
        if pd.isna(date_str) or date_str == '-':
            return None
        # 날짜 형식 파싱 (한글 포함 형식과 기타 형식 처리)
        if '년' in date_str:
            date_obj = datetime.strptime(date_str, "%Y년 %m월 %d일")
        elif '.' in date_str:
            date_obj = datetime.strptime(date_str, "%Y.%m.%d")
        else:
            date_obj = datetime.strptime(date_str, "%Y-%m-%d")
        # ISO 형식으로 변환
        return date_obj.strftime('%Y-%m-%d')
    except Exception as e:
        print(f"Error parsing date: {date_str} - {e}")
        return None

# 각 행의 Release Date를 변환 함수로 처리
filtered_releases['Release Date'] = filtered_releases['Release Date'].apply(convert_date)
filtered_releases = filtered_releases[filtered_releases['Release Date'].notna()]
filtered_releases
```

```python
# 날짜 문자열을 ISO 형식으로 변환하는 함수
def convert_date(date_str):
    formats = ['%B %d, %Y', '%Y']  # 다양한 형식 지정
    for fmt in formats:
        try:
            # 날짜 형식에 맞게 문자열 파싱
            date_obj = datetime.strptime(date_str, fmt)
            # ISO 형식(YYYY-MM-DD)으로 변환
            return date_obj.strftime('%Y-%m-%d')
        except ValueError:
            continue
    # 변환할 수 없는 경우 None 반환
    return None

filtered_releases['Date Released'] = filtered_releases['Date Released'].apply(convert_date)
filtered_releases[['Name', "Date Released"]]
```

```python
def clean_and_convert_date(date_str):
    if pd.isna(date_str):
        return None  # 결측값이 있으면 None 반환
    try:
        # 'th', 'st', 'nd', 'rd' 등의 접미사 제거
        cleaned_str = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)
        # 날짜 형식에 맞게 문자열을 파싱
        date_obj = datetime.strptime(cleaned_str, '%d %b %y')
        # ISO 형식(YYYY-MM-DD)으로 변환
        return date_obj.strftime('%Y-%m-%d')
    except Exception as e:
        # 파싱에 실패하면 None 반환
        print(f"Error parsing date: {date_str} - {e}")
        return None

# 변환 함수를 데이터프레임의 각 행에 적용
filtered_releases['release_date'] = filtered_releases['release_date'].apply(clean_and_convert_date)
filtered_releases[['title', "release_date"]]
```

```python
# `added_by_status`를 딕셔너리로 변환하고 각 상태를 컬럼으로 확장
status_columns = ['yet', 'owned', 'beaten', 'toplay', 'dropped', 'playing']

# 각 상태 컬럼을 파싱하는 함수 정의
def parse_status(row):
    try:
        status_dict = ast.literal_eval(row)
        # None은 0으로 대체, 딕셔너리에 없는 키도 0으로
        return {k: status_dict.get(k, 0) if status_dict.get(k) is not None else 0 for k in status_columns}
    except (ValueError, SyntaxError):
        # 문자열이 유효하지 않을 경우, 모든 값을 0으로 초기화
        return {k: 0 for k in status_columns}

# `added_by_status`를 파싱하고 결과를 `df`에 추가
status_df = df['added_by_status'].apply(parse_status).apply(pd.Series)
status_df.columns = [f'added_status_{col}' for col in status_columns]
df = pd.concat([df, status_df], axis=1)

df.head()
```

```python
df[df['platforms'].isna()]
```

```python
# 정규식으로 여러 구분자를 처리하기 위한 패턴 (||, |, ,)
split_pattern = r'[|,]+'

# 공백을 제거하고 플랫폼 문자열을 정리하는 함수
def clean_platforms(platforms):
    if pd.isna(platforms):
        return None
    # 각 플랫폼을 분리한 후 앞뒤 공백 제거
    platforms_cleaned = [platform.strip() for platform in re.split(split_pattern, platforms)]
    # 다시 구분자로 결합하여 반환
    return '||'.join(platforms_cleaned)

# 'platforms' 열의 모든 행에 대해 공백을 정리
df['platforms'] = df['platforms'].apply(clean_platforms)
```

```python
# 플랫폼을 담을 리스트 초기화
platforms_list = []

# 정규식으로 여러 구분자를 처리하기 위한 패턴 (||, |, ,)
split_pattern = r'[|,]+'

# NaN을 무시하고 플랫폼을 분리
for platforms in df['platforms']:
    if pd.notna(platforms):  # NaN이 아닌 경우에만 처리
        platforms_list.extend(re.split(split_pattern, platforms))

# 고유한 플랫폼 목록 추출
unique_platforms = sorted(set(platforms_list))

# 결과 확인
print("Unique Platforms:", unique_platforms)
```
